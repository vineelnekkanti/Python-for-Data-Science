#Series

It is same as NumPy array object but the  diff is series has access label, which means index by label

```python
import numpy as np
import pandas as pd
```

```python
label = ['a','b','c']
my_data=[10,20,30]
arr=np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

```python
pd.Series(data = my_data)
```

```python
#Now lets see how to label indexes while passing a LIST to Series
pd.Series(data=my_data,index=label) # now 0,1,2 changed to a,b,c
#unlike the numpy array, I can call the data points with the index name like a/b/c
```

```python
#you dont need ti write data and index. Simply put
pd.Series(my_data,label)
```

```python
#Now lets pass a NumPy array to Series
pd.Series(arr,label)
```

```python
#passing Dictionary to Series. It takes the keys as indexes
pd.Series(d)
```

A Series can hold anytype of data type as it data point

```python
pd.Series(label,label)
```

```python
s1 = pd.Series([1,2,3,4],['USA','India','Japan','UK'])
s1
```

```python
s2 = pd.Series([1,2,5,4],['USA','India','Italy','UK'])
s2
```

```python
s1['India'] #Series Indexing
```

```python
s3 = pd.Series(data=label)
s3
```

```python
s1+s2
```

If you see here the values are getting generated as float so that you dont accidentally loose information based off some division.

And also there are a couple of NaNs just because of pandas can find the same index name in the other series while adding.

**VVIMP for how to do conditional selection in PANDAS**

```python
#example:2



df[df['W']>0][['Y','X']]

#OR

result1 = df['W']>0
result2 = df[result1]
result3 = result2[['Y','X']]
result3
```

```python
df[df['W']>0].loc['D','X']
#now I have integrated the above approach with grabbing a data frame value by row index
```

```python
#giving multiple conditions
df[df['W']>0][df['X']>1]

#OR

#when you are dealing with multiple conditions, you can't use 'and' like in python
#instead of use &
df[(df['W']>0) & (df['X']>1)]
df[(df['W']>0) | (df['X']>1)]

#df[(df['W']>0) and (df['X']>1)] #This throws an ambiguity error
#df[(df['W']>0) or (df['X']>1)]
```

**MULTI-INDEX and INDEX HIERARCHY::**

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
df = pd.DataFrame(randn(6,2),hier_index,['A','B'])
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df
```

```python
df.index.names
```

```python
df.index.names = ['Groups','Numbers']
df
```

```python
#let's try to get 0.072960 out of the df
df.loc['G2'].loc[2]['B']

#OR
df.loc['G2'].loc[2,'B']
#OR

df.loc['G2',2]['B']
```

```python
#There is one more method to read the values in multi level index.
#using crossections "xs"
df.xs('G1')
```

```python
#THis will be immensly helpful when you are trying to call innner index without using outer index.
df.xs(1,level='Numbers')
```
